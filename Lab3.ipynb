# Activity 1

```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}
visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

print("Following is the Depth-First Search")
dfs(visited, graph, '5')
```

# Activity 2

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFSUtil(i, visited)

    def DFS(self):
        V = len(self.graph)
        visited = [False] * (V + 1)
        for i in range(V):
            if visited[i] == False:
                self.DFSUtil(i, visited)

g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal")
g.DFS()
```

```python

graph_tsp = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Oradea', 71)],
    'Oradea': [('Sibiu', 151)],
    'Sibiu': [('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Lugoj', 111)],
    'Lugoj': [('Mehadia', 70)],
    'Mehadia': [('Drobeta', 75)],
    'Drobeta': [('Craiova', 120)],
    'Craiova': [('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Pitesti', 97)],
    'Fagaras': [('Bucharest', 211)],
    'Pitesti': [('Bucharest', 101)],
    'Bucharest': [],
    'Giurgiu': [],
    'Urziceni': [],
    'Hirsova': [],
    'Eforie': [],
    'Vaslui': [],
    'Iasi': [],
    'Neamt': []
}

def dfs_tsp(graph, start, goal, path=[], distance=0):
    path = path + [start]

    if start == goal:
        return path, distance

    for (neighbor, weight) in graph.get(start, []):
        if neighbor not in path:
            result_path, total_dist = dfs_tsp(graph, neighbor, goal, path, distance + weight)
            if result_path:
                return result_path, total_dist
    return None, 0

start_node = 'Arad'
goal_node = 'Bucharest'
final_path, total_distance = dfs_tsp(graph_tsp, start_node, goal_node)

print(f"Path: {' -> '.join(final_path)}")
print(f"Total Distance: {total_distance}")
```

```python
import collections

graph = {
    'Arad': ['Zerind', 'Sibiu', 'Timisoara'],
    'Zerind': ['Arad', 'Oradea'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Rimnicu Vilcea': ['Sibiu', 'Craiova', 'Pitesti'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Pitesti': ['Rimnicu Vilcea', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Vaslui', 'Hirsova'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Eforie': ['Hirsova'],
    'Vaslui': ['Urziceni', 'Iasi'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Neamt': ['Iasi']
}

def bfs_count(graph, start, goal):
    visited = []
    queue = collections.deque([start])
    count = 0
    while queue:
        node = queue.popleft()
        count += 1
        if node == goal:
            return count
        if node not in visited:
            visited.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    return count

def dfs_count(graph, start, goal, visited=None, count=0):
    if visited is None:
        visited = set()
    visited.add(start)
    count += 1
    if start == goal:
        return count, True
    for neighbor in graph[start]:
        if neighbor not in visited:
            res_count, found = dfs_count(graph, neighbor, goal, visited, count)
            if found:
                return res_count, True
            count = res_count
    return count, False

# Scenario 1: BFS is smaller than DFS
start1, goal1 = 'Arad', 'Zerind'
bfs_res1 = bfs_count(graph, start1, goal1)
dfs_res1, _ = dfs_count(graph, start1, goal1)

print(f"Scenario 1 (Goal: {goal1})")
print(f"BFS nodes visited: {bfs_res1}")
print(f"DFS nodes visited: {dfs_res1}")

# Scenario 2: BFS is larger than DFS
start2, goal2 = 'Arad', 'Bucharest'
bfs_res2 = bfs_count(graph, start2, goal2)
dfs_res2, _ = dfs_count(graph, start2, goal2)

print(f"\nScenario 2 (Goal: {goal2})")
print(f"BFS nodes visited: {bfs_res2}")
print(f"DFS nodes visited: {dfs_res2}")
```
