```python
# @title Task 1

print("\n--- 1. Class Instance Namespace ---")
class Car:
    def __init__(self, name, model):
        self.name = name
        self.model = model

c1 = Car("Toyota", 2020)
print(c1.__dict__)

print("\n--- 2. Function with Optional Parameters ---")
def student_data(student_id, name=None, student_class=None):
    print("Student ID:", student_id)

    if name is not None:
        print("Name:", name)

    if student_class is not None:
        print("Class:", student_class)

student_data(101)
student_data(102, "Ali")
student_data(103, "Ahmed", "BSCS")


print("\n--- 3. Class Type, Dict Keys, Module ---")
class Student:
    pass

print(type(Student))
print(Student.__dict__.keys())
print(Student.__module__)


print("\n--- 4. Empty Classes Check ---")
class Student:
    pass

class Marks:
    pass

s = Student()
m = Marks()

print("s is Student:", isinstance(s, Student))
print("m is Marks:", isinstance(m, Marks))

print("Student inherits object:", issubclass(Student, object))
print("Marks inherits object:", issubclass(Marks, object))


print("\n--- 5. Modify Attributes ---")
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s1 = Student("Ali", 80)

print("Original:", s1.name, s1.marks)

s1.name = "Ahmed"
s1.marks = 90

print("Updated:", s1.name, s1.marks)



print("\n--- 6. Add and Delete Attributes ---")
class Student:
    def __init__(self, sid, name):
        self.sid = sid
        self.name = name

s1 = Student(1, "Ali")

s1.student_class = "BSCS"
print("Before delete:", s1.__dict__)

del s1.name
print("After delete:", s1.__dict__)


print("\n--- 7. Display All Attributes ---")
class Student:
    def __init__(self, sid, name, student_class):
        self.sid = sid
        self.name = name
        self.student_class = student_class

    def show(self):
        for key, value in self.__dict__.items():
            print(key, "=", value)

s1 = Student(1, "Ali", "BSCS")
s1.show()


print("\n--- 8. Two Instances ---")
class Student:
    def __init__(self, sid, name, marks):
        self.sid = sid
        self.name = name
        self.marks = marks

    def show(self):
        print("ID:", self.sid)
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("-----")

student1 = Student(1, "Ali", 85)
student2 = Student(2, "Ahmed", 92)

student1.show()
student2.show()


print("\n--- 9. Memory Usage ---")
import sys

class Student:
    def __init__(self, name):
        self.name = name

s1 = Student("Ali")

print("Object memory:", sys.getsizeof(s1))
print("Dict memory:", sys.getsizeof(s1.__dict__))


print("\n--- 10. Memory Optimization using __slots__ ---")
class Student:
    __slots__ = ['name']

    def __init__(self, name):
        self.name = name

s1 = Student("Ali")

print("Memory with slots:", sys.getsizeof(s1))
```

```python
# @title Task 2
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height
results = {}

data = [
    (5, 7, 4),
    (6, 8, 5),
]

for i, (a, b, h) in enumerate(data):
    t = Trapezoid(a, b, h)
    p = Parallelogram(a, h)

    t_area = t.area()
    p_area = p.area()

    results[f"Trapezoid_{i}"] = t_area
    results[f"Parallelogram_{i}"] = p_area

print("Areas:")
for shape, area in results.items():
    print(shape, "=", area)

max_shape = max(results, key=results.get)

print("\nMax Area Shape:", max_shape)
print("Area:", results[max_shape])
```
