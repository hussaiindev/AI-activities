```python
import random  # Import random module for generating random numbers

# Define the objective function (we want to maximize this)
def objective_function(x):  # Takes a single numeric input x
    return -(x ** 2) + 10  # A simple downward parabola with peak at x=0, max value=10

# Define the hill climbing algorithm
def hill_climbing(start, step_size, max_iterations):  # Takes starting point, step size, and iteration limit
    current = start  # Initialize the current solution to the starting point
    current_eval = objective_function(current)  # Evaluate the objective function at the starting point

    for i in range(max_iterations):  # Loop for a maximum number of iterations
        neighbor = current + random.choice([-1, 1]) * step_size  # Generate a neighbor by moving left or right by step_size
        neighbor_eval = objective_function(neighbor)  # Evaluate the objective function at the neighbor

        if neighbor_eval > current_eval:  # If the neighbor is better (higher value)
            current = neighbor  # Move to the neighbor
            current_eval = neighbor_eval  # Update the current evaluation
            print(f"Iteration {i+1}: Moved to {current:.4f}, Value = {current_eval:.4f}")  # Print progress
        else:  # If the neighbor is not better
            print(f"Iteration {i+1}: No improvement, staying at {current:.4f}")  # Print that we stayed

    return current, current_eval  # Return the best solution found and its value

# --- Main execution ---
start_point = random.uniform(-10, 10)  # Generate a random starting point between -10 and 10
step = 0.5  # Set the step size for generating neighbors
iterations = 50  # Set the maximum number of iterations

print(f"Starting point: {start_point:.4f}")  # Print the starting point
print(f"Starting value: {objective_function(start_point):.4f}")  # Print the value at the starting point
print("-" * 50)  # Print a separator line

best_x, best_val = hill_climbing(start_point, step, iterations)  # Run hill climbing

print("-" * 50)  # Print a separator line
print(f"Best solution: x = {best_x:.4f}")  # Print the best x found
print(f"Best value: f(x) = {best_val:.4f}")  # Print the best value found
print(f"Known optimal: x = 0, f(x) = 10")  # Print the known optimal for comparison
```
